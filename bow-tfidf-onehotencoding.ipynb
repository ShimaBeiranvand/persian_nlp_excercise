```python
import nltk

nltk.download('punkt')
from nltk.tokenize import word_tokenize

import math

from hazm import *
```

```python
# reading documents

with open('text_1.txt.txt','r', encoding="utf-8") as file:
    doc_s1 = file.read()
    
with open('text_2.txt.txt','r', encoding="utf-8") as file:
    doc_s2 = file.read()
    
with open('text_3.txt.txt','r', encoding="utf-8") as file:
    doc_s3 = file.read()
    
with open('text_4.txt.txt','r', encoding="utf-8") as file:
    doc_s4 = file.read()
    
with open('text_5.txt.txt','r', encoding="utf-8") as file:
    doc_s5 = file.read()
```

```python
print(doc_s1)
```

```python
# preprocessing

def preprocessor(doc_s):
    
    # normalizing
    normalizer = Normalizer()
    normalized_doc = normalizer.normalize(doc_s)
    
    # tokenize into words
    tokens = word_tokenize(normalized_doc)
    
    # deleting punctuations & stop words
    
    with open('persian','r', encoding="utf-8") as file:
        stop_words = file.read().split()
        
    my_punct = ['-' , '!', '"', '#', '$', '%', '&', "'", '(', ')', '*', '+', ',', '.','/', ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`', '{', '|', '}', '~', '»', '«', '“', '”' , '،']

    for token in tokens:
        if token in my_punct or token in stop_words:
            tokens.remove(token)
            
    # pos tagging
    tagger = POSTagger(model='pos_tagger.model')
    tagged_tokens = tagger.tag(tokens)
        
    # lemmatizing
    lemmatizer = Lemmatizer()

    lemmatized_tokens = []
    for token in tagged_tokens:
        lemmatized_token = lemmatizer.lemmatize(token[0])
        lemmatized_tokens.append(lemmatized_token)
        
    return lemmatized_tokens

        
doc1 = preprocessor(doc_s1)
doc2 = preprocessor(doc_s2)
doc3 = preprocessor(doc_s3)
doc4 = preprocessor(doc_s4)
doc5 = preprocessor(doc_s5)

print(doc1)
```

```python
# BAG OF WORDS TECHNIQUE
```

```python
# STEP 1 : separating unique words

all_docs_words = doc1 + doc2 + doc3 + doc4 + doc5

unique_words = []

for word in all_docs_words:
    if word not in unique_words:
        unique_words.append(word)

unique_words
```

```python
# STEP 2 : vectorization

def vectorize(doc):
    doc_vector = []
    for word in unique_words:
        word_freq = doc.count(word)
        doc_vector.append(word_freq)
    
    return doc_vector



doc1_vector = vectorize(doc1)
doc2_vector = vectorize(doc2)
doc3_vector = vectorize(doc3)
doc4_vector = vectorize(doc4)
doc5_vector = vectorize(doc5)


print(doc1_vector)

```

```python
# TF_IDF

def tf_idf(words , total_docs):
    
    
    tf_idf = dict([])
    
    for word in words:
        print('word  ', word)
        
        # tf
        word_freq = 0
        for other_word in words:
            if other_word == word:
                word_freq += 1
       # word_freq = words.count(word)
        tf = word_freq  / len(words)
        print(tf)
        
        # idf
        doc_freq = 0
        for the_doc in total_docs:
            if word in the_doc:
                doc_freq += 1
        idf = math.log(5 / doc_freq )
        print(idf)
        
        # tf_idf
        tf_idf[word] = tf * idf
        print(tf_idf[word])
        
    return tf_idf



total_docs = [doc1 , doc2 , doc3 , doc4 , doc5]
words_tf_idf = tf_idf(unique_words , total_docs)

print(words_tf_idf)


```

```python
# N_GRAMS
```

```python
# creating a dictionary including 1gram , 2gram & 3gram for every doc


grams_dict = dict([])

for doc_num , doc_s in zip(range(1,6) , [doc_s1 , doc_s2 , doc_s3 , doc_s4 , doc_s5]):
    for n in range(1,4):
        
        key = 'doc' + str(doc_num) + '_' + str(n) + 'gram'
        
        grams_dict[key] = nltk.ngrams(word_tokenize(doc_s) , n)
        
        # coverting n_grams(which are in zip type) to list
        listed_value = []
        for i in grams_dict[key]:
            listed_value.append(i)
            
        grams_dict[key] = listed_value
        
        
print(grams_dict)
```

```python
# TF_IDF for grams in a dictionary

grams_tf_idf = dict([])

for doc_gram in grams_dict.keys():
    
    print('doc_gram: ',doc_gram)
    
    key = doc_gram + '_tf_idf'
    
    print('key: ',key)
    
    if '1gram' in key :
        total_docs = [grams_dict['doc1_1gram'] , grams_dict['doc2_1gram'] , grams_dict['doc3_1gram'] , grams_dict['doc4_1gram'] , grams_dict['doc5_1gram']]
        
    elif '2gram' in key :
        total_docs = [grams_dict['doc1_2gram'] , grams_dict['doc2_2gram'] , grams_dict['doc3_2gram'] , grams_dict['doc4_2gram'] , grams_dict['doc5_2gram']]
            
    else :
        total_docs = [grams_dict['doc1_3gram'] , grams_dict['doc2_3gram'] , grams_dict['doc3_3gram'] , grams_dict['doc4_3gram'] , grams_dict['doc5_3gram']]
                
    
    print('grams_dict[doc_gram] : ',grams_dict[doc_gram])
    
    grams_tf_idf[key] = tf_idf(grams_dict[doc_gram] , total_docs)
    
    print('grams_tf_idf[key] : ',grams_tf_idf[key])
    
    
```
